```python
import os
import tensorflow as tf
import cv2
import numpy as np
import seaborn as sns
import gc
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Main directory where model data is saved
mainDataDirectory = "C:/Users/seyed/Music/Brain Tumor - Model Data/"
# Categories of image classification
imageCategories = ["No", "Yes"]

# Directorie for testing data
testDirectory = os.path.join(mainDataDirectory, 'Test')
```

```python
# Function to read images from a directory and add labels to each image
def read_images_from_directory(directory):
    images = []
    labels = []
    for imageCategory in imageCategories:
        category_path = os.path.join(directory, imageCategory)
        label = imageCategories.index(imageCategory)
        
        for filename in os.listdir(category_path):
            image_path = os.path.join(category_path, filename)
            image = cv2.imread(image_path)
            if image is not None:
                images.append(image)
                labels.append(imageCategories[label])
            else:
                print(f"Error in reading the image: {image_path}")
    return images, labels
```

```python
test_images, test_labels = read_images_from_directory(testDirectory)
```

```python
## VGG 16 model for Tumour detection model testing

# Load your trained model
model_Directory = 'C:/Users/seyed/Music/Brain Tumor - Model/TumorDetectionModel_VGG-16.h5'
model = tf.keras.models.load_model(model_Directory)

# Test your model on the test data
predicted_labels_vgg16 = []

for test_image in test_images:
    # Preprocess the test image
    # test_image = test_image / 255.0  # Normalize pixel values to be between 0 and 1
    test_image = np.expand_dims(test_image, axis=0)  # Add batch dimension

    # Get the model predictions39
    predictions = model.predict(test_image)
    predicted_class = np.argmax(predictions)
    predicted_labels_vgg16.append(imageCategories[predicted_class])
```

```python
# Create a confusion matrix
conf_matrix = confusion_matrix(test_labels, predicted_labels_vgg16)

# Plot the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=imageCategories, yticklabels=imageCategories)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()

# Print classification report for additional metrics
class_report = classification_report(test_labels, predicted_labels_vgg16)
print("Classification Report:\n", class_report)
```
